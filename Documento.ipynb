```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
arquivoEntrada = open(r'C:\Users\Uchoa\Desktop\Fatec\Dados\NASCIDOS_VIVOS_1.csv')
arquivoSaida = open(r'C:\Users\Uchoa\Desktop\Fatec\Dados\NASCIDOS_VIVOS_NOVOS.csv', 'w', newline = '')
escrever = csv.writer(arquivoSaida)

 
while True:
    linhaNova = []
    linha = arquivoEntrada.readline()
    
    if linha == [''] or linha == '':
        break
    
    if len(linha) != 51 and len(linha) != 123:
        continue
    
    linha = linha.split(';')
    if (linha.count('') > 0):
        linha.remove('')
        
    if (linha.count('\n') > 0):
        linha.remove('\n')
        
    for dado in linha:
        dado.strip('"')
        linhaNova.append(dado.strip())

 

    escrever.writerow(linhaNova)
    linhaNova = []
        
arquivoEntrada.close()
arquivoSaida.close()
 
```
